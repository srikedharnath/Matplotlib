```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\ADMIN\Downloads\Practice-Datasets .csv")
```

```python
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'])
df.head(4)
```

1. Create a histogram to analyze the distribution of Age.

```python
plt.hist(df['Age'], bins=15, edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

# 2. Plot a pie chart showing the percentage distribution of Gender.

```python
df['Gender'].value_counts().plot(
    kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()
```

```python
# 3. Create a bar chart showing the number of customers in each City.
df['City'].value_counts().plot(kind='bar')
plt.title("Customers in Each City")
plt.xlabel("City")
plt.ylabel("Count")
plt.show()
```

```python
# 4. Plot a box plot to analyze Monthly Income distribution.
plt.boxplot(df['Monthly_Income'])
plt.title("Monthly_Income_Distribution")
plt.show()
```

```python
# 5. Create a histogram of Spending Score and interpret the skewness.
plt.hist(df['Spending_Score'], bins=15, edgecolor='black')
plt.title("Spending_Score Distribution")
plt.show()
print("Skewness:", df['Spending_Score'].skew())
```

```python

# 6. Plot a line chart to visualize Purchase Amount trend over Purchase Date.
df.sort_values('Purchase_Date').plot(x='Purchase_Date', y='Purchase_Amount', kind='line')
plt.title("Purchase_Trend_Over_Time")
plt.show()
```

```python

# 7. Create a bar plot to compare average Monthly Income by Gender.
df.groupby('Gender')['Monthly_Income'].mean().plot(kind='bar')
plt.title("Average Monthly Income by Gender")
plt.show()
```

```python

# 8. Plot a box plot of Purchase Amount grouped by Product Category.
df.boxplot(column='Purchase_Amount', by='Product_Category')
plt.title("Purchase Amount by Product Category")
plt.suptitle("")
plt.show()
```

```python
# 9. Create a pie chart to show Product Category distribution.
df['Product_Category'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Product Category Distribution")
plt.ylabel("")
plt.show()
```

```python
# 10. Plot a histogram of Purchase Amount with appropriate bin size.
plt.hist(df['Purchase_Amount'], bins=20, edgecolor='black')
plt.title("Purchase Amount Distribution")
plt.show()
```

```python
# 11. Create a bar chart showing average Spending Score by City.
df.groupby('City')['Spending_Score'].mean().plot(kind='bar')
plt.title("Average Spending Score by City")
plt.show()
```

```python
# 12. Plot a line chart showing total Purchase Amount per month.
df["Month"]=pd.to_datetime(df["Purchase_Date"]).dt.month
monthly = df.groupby('Month')['Purchase_Amount'].sum()
monthly.plot(kind='line',marker="o")
plt.title("Total Purchase Amount per Month")
plt.show()
```

```python

# 13. Create a box plot comparing Monthly Income across different Cities.
df.boxplot(column='Monthly_Income', by='City')
plt.title("Monthly Income by City")
plt.suptitle("")
plt.show()
```

```python

# 14. Plot a histogram to analyze Age distribution by Gender.
for gender in df['Gender'].unique():
    plt.hist(df[df['Gender'] == gender]['Age'], bins=10, alpha=0.5, label=gender)

plt.legend()
plt.title("Age Distribution by Gender")
plt.show()
```

```python
# 15. Create a bar chart showing total Purchase Amount by Product Category.
df.groupby('Product_Category')['Purchase_Amount'].sum().plot(kind='bar')
plt.title("Total Purchase Amount by Product Category")
plt.show()
```

```python
# 16. Plot a scatter plot between Monthly Income and Spending Score.
plt.scatter(df['Monthly_Income'], df['Spending_Score'], alpha=0.6)
plt.xlabel("Monthly_Income")
plt.ylabel("Spending_Score")
plt.title("Income vs Spending Score")
plt.show()
```

```python


# 17. Create a line chart to visualize daily Purchase Amount variation.
daily = df.resample('D', on='Purchase_Date')['Purchase_Amount'].sum()
daily.plot(kind='line')
plt.title("Daily Purchase Amount")
plt.show()
```

```python
# 18. Plot a box plot of Spending Score grouped by Gender.
df.boxplot(column='Spending_Score', by='Gender')
plt.title("Spending Score by Gender")
plt.suptitle("")
plt.show()
```

```python

# 19. Create a bar chart showing customer count by Product Category and Gender.
pd.crosstab(df['Product_Category'], df['Gender']).plot(kind='bar')
plt.title("Customer Count by Product Category and Gender")
plt.show()
```

```python

# 20. Analyze the relationship between Age and Purchase Amount using a suitable visualization.
plt.scatter(df['Age'], df['Purchase_Amount'], alpha=0.6)
plt.xlabel("Age")
plt.ylabel("Purchase_Amount")
plt.title("Age vs Purchase Amount")
plt.show()
```
